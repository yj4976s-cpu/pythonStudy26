```python
#숫자형(Numbertype) : 숫자로 이루어진 자료형으로
# 123과 같은 정수, 12.34와 같은, 실수, 8진법,  16진법 등이다.

# 정수형(integer) : 양의정수, 음의정수, 0
a = 123
a = -178
a = 0
```

```python
# 실수형(floating-point) : 소수점이 포함된 숫자
a = 1.2
a = -3.45
```

```python
# 컴퓨터식 지수 표현 방식 : 4.24e10 또는 4.24E10처럼 표현한다(e와 E 둘 중 어느 것도 사용가능)
a = 4.24E10
a = 4.24e-10
#여기서 4.24E10은 4.24 * 10^10, 4.24e-10은 4.24 * 10^-10을 의미한다.
```

```python
# 8진수(octal)를 만들기 위해서는 숫자가 0o또는0O로 시작하면 된다
a = 0o177
print(a)
# 0o177 = 1*8^2 + 7*8^1 + 7 = 127
```

```python
#16진수(hexadecimal)를 만들기 위해서는 0x로 시작하면 된다.
a = 0x8ff
b = 0xABC
print(b)
# 0xABC = 10*16^2 + 11*16^1 + 12 = 2748
```

```python
#사칙연산
a = 3
b = 4
print(a+b)

print(a-b)

print(a*b)
print(a/b)
```

```python
# x의 y제곱을 나타내는 ** 연산자
a = 3
b = 4
a ** b
```

```python
# 나눗셈 후 나머지를 반환하는 % 연산자
7 % 3
3 % 7
```

```python
# 나눗셈 후 몫을 반환하는 // 연산자
7 // 4
```

```python
# 복합 연산자는 산술 연산자(+,-)와 대입연산자(=)를 합쳐놓은 것이다
a = 1
a = a + 1
print(a)
a += 1
a -= 1
print(a)
```

```python
a = 1
a -= 1
print(a)
```
