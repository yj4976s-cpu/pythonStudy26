```python


# if문은 조건문으로 판단하여 참(True) 거짓(False)로 판단하는 코드
# if 조건 :
#     조건이 참일 때 실행하는 문
# elif 조건 :
#     위 조건이 False일 때 또 비교하는 조건문
# else :
#      위 조건이 모두 해당하지 않을때 실행하는 조건문

# 학점 계산용 코드
avg = int(input("평균을 입력하세요 >>>"))
if avg >= 90 :
   grade = "A" # 90점이상이면 A

elif avg >= 80 :
    grade = "B" # 80점이상이면 B

elif avg >= 70 :
    grade = "C" # 70점이상이면 C

else:
    grade = "F"

print("귀하의 등급은 : " + grade + "입니다.")
```

```python
marks = [90, 25, 67, 45, 80]   # 학생들의 시험 점수 리스트

number = 0   # 학생에게 붙여 줄 번호
for mark in marks:   # 90, 25, 67, 45, 80을 순서대로 mark에 대입
    number = number +1
    if mark >= 60:
        print("%d번 학생은 합격입니다." % number)
    else:
        print("%d번 학생은 불합격입니다." % number)
```

```python
marks = [90, 25, 67, 45, 80]
for number in range(len(marks)):
    if marks[number] < 60:
        continue
    print("%d번 학생 축하합니다. 합격입니다." % (number+1))
```

```python
for i in range(2, 10):
    for j in range(1, 10):
        print(i*j, end=" ") # 이와 같이 print 함수에 end 파라미터를 설정한 이유는 해당 결괏값을 출력할 때 다음 줄로 넘기지 않고 그 줄에 계속 출력하기 위해서이다.
    print('') # 2단, 3단 등을 구분하기 위해 사용한다. 두 번째 for 문이 끝나면 결괏값을 다음 줄부터 출력하게 하는 역할을 한다.
```

```python
a = [1, 2, 3, 4] # 리스트안에 for문을 포함하는 리스트 컴프리헨션(comprehension)
result = [num * 3 for num in a]
print(result)
```

```python
a = [1, 2, 3, 4] # 리스트 컴프리헨션의 문법
result = [num * 3 for num in a if num % 2 == 0] # [표현식 for 항목 in 반복_가능_객체 if 조건문]
print(result) # for문을 여려개 사용할 때의 문법
            # [표현식 for 항목1 in 반복_가능_객체1 if 조건문1
                  # for 항목2 in 반복_가능_객체2 if 조건문2
                  # ...
                  # for 항목n in 반복_가능_객체n if 조건문n]
```

```python
fruits = ['apple', 'banana', 'orange']
for i, fruit in enumerate(fruits, 1): # 1부터 시작
    print(f"{i}: {fruit}")
```

```python
names = ['홍길동', '김철수', '이영희'] # 두 개 이상의 리스트를 동시헤 순회하고 있을때 zip 함수 사용!!
scores = [85, 92, 78]
for name, score in zip(names, scores):
    print(f"{name}: {score}점")
```

```python
names = ['홍길동', '김철수', '이영희']
korean = [85, 92, 78]
english = [90, 88, 95]
for name, korean, english in zip(names, korean, english):
    print(f"{name}: {korean}점: {english}점")
```
