```python
# 집합 자료형(set)
# 로또 주머니와 같이 중복을 허용하지 않고 순서가 없는
# 데이터들의 모임이다.
# 교집합, 합집합, 차집합 등의 집합 연산을 쉽계 처리

s1 = set([1, 2, 3]) # set()에 값을 넣으면 된다.
print(s1)
```
