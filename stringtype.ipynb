```python
# 주민번호 앞 2자리를 생년으로 추출 year
# 주민번호 3~4 자리를 생월로 추출 month
# 주민번호 5~6 자리를 생일로 추출 day
# 주민전호 8번째 1자리를 남여 추출 sex
inputSsn = "120604-2345678"
year = inputSsn[0:2]
month = inputSsn[2:4]
day = inputSsn[4:6]
print(" 귀하의 생년은 " + year + "년")
print(" 귀하의 생월은 " + month + "월")
print(" 귀하의 생일은 " + day + "일 입니다.")

# 주민번호를 이용해서 남여구분을 위한 코드
# 문자열로 인식된 숫자를 숫자로 계산시 타입변환 필요
ssnNumber = inputSsn[7]
# print(ssnNumber)
# type(ssnNumber)
ssnNumber = int(ssnNumber) # 문자가 숫자로 타입변환진행 int(문자)

if ssnNumber % 2 == 1 : # 숫자로된 성별을 2로 나눈 나머지가 1
    sex = "남자"
else:
    sex = "여자"
print(sex)
# 미션 주민번호 8번째 글자가 1,2,5,6이면 1900년 생
# 3, 4, 7, 8이면 2000년 생으로 만들어보자
```

```python
if ssnNumber == 1 :
    year = "19"+year # 문자인경우 + 기호는 붙임
elif ssnNumber == 2 :
    year = "19"+year
elif ssnNumber == 5 :
    year = "19"+year
elif ssnNumber == 6 :
    year = "19"+year
elif ssnNumber == 9 :
    print("당신은 외계인입니다. 돌아가세요!!")
else:
    year = "20"+year

print(year)
```

```python
s = "Python"
s.isalpha()
s ="Python3"
s.isalpha()
s = "hello world"
s.isalpha()
# isalpha 함수는 문자열이 알파벳 문자로만 이루어져 있는지 검사한다. 영문 알파벳만 포함되어 있는경우 True를 반환, 공백이나 숫자, 특수 문자가 포함되어 있는 경우 False를 반환한다.
```

```python
s = "12345"
s.isdigit()
s = "1234a"
s.isdigit()
s = "12 34"
s.isdigit()
# isdigit 함수는 문자열이 숫자로만 이루어져 있는지를 검사한다. 문자열의 모든 문자가 숫자일 경우 True, 하나라도 숫자가 아닌 문자가 포함되어 있으면 False를 반환한다.
```

```python
s =  "Life is too short"
s.startswith("Life")
s.startswith("short")
# startswith 함수는 문자열이 지정한 문자(열)로 시작하는지를 검사한다. 지정한 문자열로 시작하면 True를 반환, 그렇지 않으면 False를 반환한다.
```

```python
s = "Life is too short"
s.endswith("short")
s.endswith("too")
# endswith 함수는 문자열이 지정한 문자(열)로 끝나는지를 검사한다. 지정한 문자열로 끝나면 True를 반환, 그렇지 않으면 False를 반환한다.
```
